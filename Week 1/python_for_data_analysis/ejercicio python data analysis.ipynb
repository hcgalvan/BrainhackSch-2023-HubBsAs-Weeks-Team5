```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
```

```python
print (iris.keys())
```

```python
print(iris['data'].shape)
print(type(iris['data']))
```

```python
data = iris['data']
feature_names = iris['feature_names']
```

```python
df = pd.DataFrame(data, columns=feature_names)

```

```python
print(df.describe())
```

```python
subset_df = df.head(50)
```

```python
def find_extreme_values(column):
    mean = column.mean()
    std = column.std()
    extreme_values = [value for value in column if abs((value - mean) / std) > 3.9]
    return extreme_values

extreme_sepal_lengths = find_extreme_values(df['sepal length (cm)'])
print(extreme_sepal_lengths)
```

```python
def find_extreme_values(column):
    mean = column.mean()
    std = column.std()
    extreme_values = [value for value in column if abs((value - mean) / std) > 3.9]
    return extreme_values

def find_extreme_values_in_dataframe(df):
    extreme_values = {}
    for column in df.columns:
        extreme_values[column] = find_extreme_values(df[column])
    return extreme_values

extreme_values = find_extreme_values_in_dataframe(df)
print(extreme_values)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 6))
for i, column in enumerate(df.columns):
    axes[i].boxplot(df[column])
    axes[i].set_title(column)
    axes[i].set_ylabel('cm')
plt.suptitle('Boxplot Distribution for Iris Features')
plt.show()
```

```python
df.to_csv('iris_dataframe.csv', index=False)

fig.savefig('iris_boxplot.png')
```
